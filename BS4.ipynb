```python
import requests as req
import lxml
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/w/index.php'
res = req.post(url,data = {'search':input('enter which you want to search :')})
file = open('sdata.html','w')
file.write(res.text)
file.close()
```

```python
soup = BeautifulSoup(open('sdata.html','r').read(),'lxml')
```

```python
for i in soup.findAll('span'):
    try:
        if(i['class']=='toctext'):
            print(i.text)
    except:
        pass
```

```python
for i in soup.findAll('li'):
    try:
        if('toc' in i['class'][0]):
            print(i.text)
           
    except:
        pass
   
```
