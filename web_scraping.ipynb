```python
import requests as req


url = 'https://www.google.co.in/search?q=hello&oq=hello&aqs=chrome..69i57j0l5.707j0j7&sourceid=chrome&ie=UTF-8'

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}

res = req.get(url,headers = headers)
```

```python
page = res.text
```

```python
import re

k = re.finditer('<cite class="iUh30">',page)
```

```python
for i in k:
    t = i.start()
    #print(page[t:t+100])
    r =page[t:t+100]
    q=r.split('>')[1][:-6]
    print (q)
```
